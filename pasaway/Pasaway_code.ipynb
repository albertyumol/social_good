```python
import numpy as np
import matplotlib.pyplot as plt  
from matplotlib.animation import FuncAnimation, PillowWriter  
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
from matplotlib.dates import DateFormatter

%matplotlib notebook  
```

```python
x_d = df_ph_manila['date1']

y = df_ph_manila['retail_and_recreation_percent_change_from_baseline']
y1 = df_ph_manila['grocery_and_pharmacy_percent_change_from_baseline']
y2 = df_ph_manila['parks_percent_change_from_baseline']
y3 = df_ph_manila['transit_stations_percent_change_from_baseline']
y4 = df_ph_manila['workplaces_percent_change_from_baseline']
y5 = df_ph_manila['residential_percent_change_from_baseline']

fig, ax = plt.subplots(figsize=(10,5), dpi=500)  
x, y_, y1_, y2_, y3_, y4_, y5_ = [], [], [], [], [], [], []  
ln1, = plt.plot([], [], color='r', label= 'Retail and Recreation')  
ln2, = plt.plot([], [], color='b', label= 'Grocery and Pharmacy')
ln3, = plt.plot([], [], color='g', label= 'Parks')  
ln4, = plt.plot([], [], color='c', label= 'Transit Stations')
ln5, = plt.plot([], [], color='m', label= 'Workplaces') 
ln6, = plt.plot([], [], color='y', label= 'Residential')


import datetime as dt


plt.title('Mobility of Filipinos in target locations during various quarantine periods (Metro Manila)')
plt.ylabel('% change compared to baseline')
plt.axhline(0, color = 'k')
plt.axvline(dt.datetime(2020, 3, 15), color = 'k', linestyle = '--', alpha=0.5)
plt.axvline(dt.datetime(2020, 6, 1), color = 'k', linestyle = '--', alpha=0.5)
plt.axvspan(dt.datetime(2020, 3, 15), dt.datetime(2020, 6, 1), ymin=0, ymax=0.625, alpha=0.3, color='green')
# plt.legend()
plt.text(dt.datetime(2020, 3, 18),-10,'ECQ Period', fontsize=10, style='italic')
plt.text(dt.datetime(2020, 2, 19),-52,'Workplaces', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 2, 26),20,'Parks', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 5, 14),45,'Residential', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 5, 1),-35,'Grocery and Pharmacy', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 4, 15),-89,'Retail', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 3, 17),-97,'Transit Stations', fontsize=9, fontweight='bold')
plt.text(dt.datetime(2020, 3, 8),55,'Start of', fontsize=7)
plt.text(dt.datetime(2020, 3, 8),51,'ECQ', fontsize=7)
plt.text(dt.datetime(2020, 6, 2),55,'Start of', fontsize=7)
plt.text(dt.datetime(2020, 6, 2),51,'GCQ', fontsize=7)


def init():  
    ax.set_xlim(0, len(x_d))  
    ax.set_ylim(-100, 60)
    ax.set_xlim(min(x_d), max(x_d))  
    ax.set_ylim(-100, 60)
    date_form = DateFormatter("%b-%d")
    ax.xaxis.set_major_formatter(date_form)
  
def update(i):  
    x.append(x_d.iloc[int(i)]) 
    y_.append(y.iloc[int(i)])  
    y1_.append(y1.iloc[int(i)])
    y2_.append(y2.iloc[int(i)]) 
    y3_.append(y3.iloc[int(i)]) 
    y4_.append(y4.iloc[int(i)]) 
    y5_.append(y5.iloc[int(i)]) 
    ln1.set_data(x, y_)  
    ln2.set_data(x, y1_) 
    ln3.set_data(x, y2_)  
    ln4.set_data(x, y3_) 
    ln5.set_data(x, y4_)  
    ln6.set_data(x, y5_) 
    

ani = FuncAnimation(fig, update, np.linspace(0, len(x_d), len(x_d)+1), init_func=init, repeat = True, blit = True)
plt.show()


# writer = PillowWriter(fps=25)  
# ani.save("demo.gif", writer=writer)  
```
